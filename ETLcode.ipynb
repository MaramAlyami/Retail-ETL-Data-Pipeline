```python
df_clean = ecommerce_df.copy()

# Clean the string columns: remove leading/trailing spaces and convert to lowercase
df_clean[df_clean.select_dtypes(include='object').columns] = \
    df_clean.select_dtypes(include='object').apply(lambda col: col.str.strip().str.lower())

# Check and remove duplicates in one step
num_duplicates = df_clean.duplicated().sum()

if num_duplicates > 0:
    print(f' Number of duplicate rows: {num_duplicates}')
    print('\n Duplicate rows:')
    display(df_clean[df_clean.duplicated()])

    # Remove duplicates
    df_clean = df_clean.drop_duplicates()
    print(f'\nDuplicates removed. Number of rows now: {df_clean.shape[0]}')
else:
    print(' No duplicate rows.')

# Verify that no duplicates remain after removal
print(f'Number of duplicate rows after removal: {df_clean.duplicated().sum()}')
```

```python
df_clean = df_clean.drop_duplicates()
```

```python
def fill_missing_with_default(df, column, default_value=0):
    """
    Fill missing values in a column with a default value (e.g., 0).
    """
    df[column] = df[column].fillna(default_value)
    return df
```

```python
df_clean = fill_missing_with_default(df_clean, 'Discount', default_value=0)
```

```python

ecommerce_df['Order Date'] = ecommerce_df['Order Date'].astype(str)


ecommerce_df['Order Date'] = ecommerce_df['Order Date'].str.strip("'\"")


ecommerce_df['Order Date'] = pd.to_datetime(ecommerce_df['Order Date'], errors='coerce')


print(ecommerce_df['Order Date'].head())

```

```python
ecommerce_df = ecommerce_df.dropna(subset=['Customer Name'])
```

```python
total_revenue = df_clean['Sales'].sum()
print(f'Total Revenue: {total_revenue}')
```

```python
average_sales = df_clean['Sales'].mean()
print(f'Average Sales per Order: {average_sales}')
```

```python
df_clean['Discount Percentage'] = df_clean['Discount'] * 100
df_clean
```

```python

def categorize_customer(row):
    if row['Sales'] > 1000:
        return 'High Value'
    elif row['Sales'] > 500:
        return 'Medium Value'
    else:
        return 'Low Value'

df_clean['Customer Segment'] = df_clean.apply(categorize_customer, axis=1)
df_clean
```
